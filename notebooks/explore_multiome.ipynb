```python
with open('genes.txt', 'r') as f:
    genes = f.readlines()
    
genes = [g.strip() for g in genes]
```

```python
genes[:10]
```

```python
multiome_gene_count = 0
multiome_genes = set(df['location'])
overlap_genes = []
for gene in genes:
    if gene in multiome_genes:
        overlap_genes.append(gene)
        multiome_gene_count += 1

len(overlap_genes)
```

```python
overlap_genes[:10]
```

```python
df[df['location'] == 'A1BG-AS1']
```

```python
from collections import defaultdict

overlap_genes_set = set(overlap_genes)
```

```python
from tqdm import tqdm
import pandas as pd

filtered_df = df[df['location'].isin(overlap_genes_set)]

gene_to_average_counts = filtered_df.groupby('location')['normalized_count'].mean().to_dict()
```

```python
gene_to_average_counts.values()
```

```python
import numpy as np

values = np.array(list(gene_to_average_counts.values()))
mean = values.mean()
std = values.std()

# Normalize the values
normalized_gene_to_average_counts = {gene: (count - mean) / std for gene, count in gene_to_average_counts.items()}
```

```python
normalized_gene_to_average_counts
```

```python
mean, std
```

```python
normalized_counts = list(normalized_gene_to_average_counts.values())
len(normalized_counts)
```

```python
all_gene_to_counts = {}
for gene in genes:
    if gene in normalized_gene_to_average_counts:
        all_gene_to_counts[gene] = normalized_gene_to_average_counts[gene]
    else:
        all_gene_to_counts[gene] = 0
```

```python
all_gene_to_counts
```

```python
import pickle

with open('relative_accessibility.pkl', 'wb') as f:
    pickle.dump(all_gene_to_counts, f)
```
